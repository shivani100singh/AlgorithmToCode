````python
dataset = datasets.load_dataset('csv', data_files='data.csv', delimiter=',')
````

````python
dataset = dataset['train']
````

````python
import numpy as np

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")

tokenized_algo = tokenizer(dataset["algorithm"], return_tensors="np", padding=True)
tokenized_algo = dict(tokenized_algo)

tokenized_code = tokenizer(dataset["code"], return_tensors="np", padding=True)
tokenized_code = dict(tokenized_code)
````

````python
tokenized_algo
````

````python
from transformers import TFAutoModelForSeq2SeqLM
from tensorflow.keras.optimizers import Adam

model = TFAutoModelForSeq2SeqLM.from_pretrained("t5-small")
model.compile(optimizer=Adam(3e-5))
````

````python
import tensorflow

tensorflow.debugging.set_log_device_placement(True)

tokenized_algo = np.asarray([np.asarray(l) for l in np.asarray(tokenized_algo['input_ids'])])
tokenized_code = np.asarray([np.asarray(l) for l in np.asarray(tokenized_code['input_ids'])])

model.fit(np.asarray(tokenized_algo), np.asarray(tokenized_code), epochs=50)
````

````python
prediction_string = tokenizer.encode('''//Add Two numbers
step 1: store one number in variable a.
Step 2: Store another number in variable b.
Step 3: Add a and b.
Step 4: Store the result of step 3 in variable sum.
Step 5: print sum.''', return_tensors='np', padding=True)

output_ids = model.generate(prediction_string, max_length=50, num_beams=4, early_stopping=True)
````
